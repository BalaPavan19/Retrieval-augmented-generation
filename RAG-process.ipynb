```python
import PyPDF2

def extract_text_from_pdf(file_path):
    with open(file_path, 'rb') as file:
        reader = PyPDF2.PdfReader(file)
        text = ''
        for page in range(len(reader.pages)):
            text += reader.pages[page].extract_text()
        return text

# Example usage
file_path = '/content/drive/My Drive/Data/BOI-notes-II-sem-B.Com-2019-Batch-onwards.pdf'
pdf_text = extract_text_from_pdf(file_path)
print(pdf_text)
```

```python
def extract_text_from_txt(file_path):
    with open(file_path, 'r') as file:
        return file.read()

file_path = '/content/drive/My Drive/your_folder/example.txt'
txt_text = extract_text_from_txt(file_path)
print(txt_text)
```

Text Embedding for Retrieval (Using BERT or Similar Models)

```python
!pip install transformers
```

```python
from transformers import BertTokenizer, BertModel
import torch

# Load pre-trained BERT model
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
model = BertModel.from_pretrained('bert-base-uncased')

def get_embeddings(text):
    inputs = tokenizer(text, return_tensors='pt', truncation=True, padding=True, max_length=512)
    outputs = model(**inputs)
    embeddings = outputs.last_hidden_state.mean(dim=1)  # Mean of the token embeddings
    return embeddings

# Get embeddings for all documents
doc_embeddings = [get_embeddings(doc) for doc in documents]
```

Build the Retrieval Function : Cosine Similarity Function

```python
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

def retrieve(query, doc_embeddings, documents):
    # Get embedding for the query
    query_embedding = get_embeddings(query)

    # Convert the embeddings to numpy arrays and ensure they are 2D
    query_embedding = query_embedding.mean(dim=1).detach().numpy().reshape(1, -1)  # Ensuring it's 2D
    doc_embeddings = np.array([emb.mean(dim=1).detach().numpy() for emb in doc_embeddings])

    # Compute cosine similarity between the query and document embeddings
    similarities = cosine_similarity(query_embedding, doc_embeddings)

    # Get the document with the highest similarity score
    best_doc_index = similarities.argmax()
    return documents[best_doc_index]

# Example query
query = "Essentials of a valid Endorsement"
best_doc = retrieve(query, doc_embeddings, documents)
print(best_doc)
```

Set Up the Generative Model (GPT-2 or GPT-3 via Hugging Face)

```python
!pip install transformers
```

```python
from transformers import GPT2Tokenizer, GPT2LMHeadModel

# Load pre-trained GPT-2 model
gpt_tokenizer = GPT2Tokenizer.from_pretrained('gpt2')
gpt_model = GPT2LMHeadModel.from_pretrained('gpt2')

def generate_response(prompt):
    inputs = gpt_tokenizer.encode(prompt, return_tensors='pt', max_length=512, truncation=True)
    # Set max_new_tokens to limit the number of tokens generated
    outputs = gpt_model.generate(inputs, max_new_tokens=100, num_return_sequences=1)
    return gpt_tokenizer.decode(outputs[0], skip_special_tokens=True)

# Use the retrieved document as the prompt
response = generate_response(best_doc)
print(response)
```
